# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # creating subplots and formating figures
import seaborn as sns # visualizations including heatmaps
%matplotlib inline
```

# Importing the Dataset

```python
df = pd.read_csv(r"C:\Users\sraja\Downloads\sales_data_sample.csv")
df.tail()
```

# Bar Plot



```python
sns.barplot(x ='STATUS',y='SALES', data = df)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x ='STATUS',y='SALES', hue = 'PRODUCTLINE', data = df)
```

# Horizontal Barplot

```python
sns.barplot(x ='SALES',y='STATUS',capsize= 0.2, data = df)
```

# Lineplot

```python
sns.lineplot(x='MONTH_ID', y= 'SALES', data=df)
```

## Without Confidence Interval

```python
sns.lineplot(x='MONTH_ID', y= 'SALES', ci = None, data=df)
```

## Tweaking the CI

```python
sns.lineplot(x='MONTH_ID', y= 'SALES', ci = 'sd', data=df)
```

## With estimator

```python
sns.lineplot(x='MONTH_ID', y= 'SALES', ci = None, estimator= sum, data=df)
```

```python
sns.lineplot(x='MONTH_ID', y= 'SALES', ci = None, estimator= None , data=df)
```

## Grouping the Lineplot

```python
sns.lineplot(x='MONTH_ID', y= 'SALES', ci = None, hue= 'YEAR_ID', data=df)
```

```python
sns.lineplot(x='MONTH_ID', y= 'SALES', ci = None, hue= 'YEAR_ID',palette = 'dark:salmon_r', data=df)
```

# Boxplot

```python
sns.boxplot(x='STATUS', y='SALES', data=df)
```

```python
sns.boxplot(x='STATUS', y='SALES', showmeans= True, data=df)
```

# Histogram

```python
sns.histplot(x='SALES',data= df)
```

## Binwidth

```python
sns.histplot(x='SALES',binwidth=600,data= df)
```

## Adjusting the Histogram

```python
sns.histplot(x='SALES',bins= np.arange(0,14000,1000), data= df)
```

## Frequency Line

```python
sns.histplot(x='SALES',bins= np.arange(0,14000,1000),kde= True, data= df)
```

## Polyplot Histogram

```python
sns.histplot(x='SALES',element ='poly',data= df)
```

```python
sns.histplot(x='SALES',stat= 'probability', data= df)
```

# Scatterplot

```python
plt.figure(figsize=(10, 10))
sns.scatterplot(x='SALES',y= 'PRICEEACH', data= df)
```

## Grouping the Scatterplot

```python
sns.scatterplot(x='SALES',y= 'PRICEEACH', hue= 'YEAR_ID', data= df)
```

## With Markers

```python
plt.figure(figsize=(10, 10))
sns.scatterplot(x='SALES',y= 'PRICEEACH', style = 'YEAR_ID', size= 'ORDERLINENUMBER', data= df)
```

# Heatmap


```python
df2= df.pivot_table(index='STATUS',columns='PRODUCTLINE',values='SALES')
df2
df2.fillna(0)
```

```python
sns.set_style('white')
plt.figure(figsize=(5,5))
sns.heatmap(df2)
```

# With Numbers in cell

```python
sns.heatmap(df2,annot=True)
```

# Integers

```python
sns.heatmap(df2,annot=True,fmt='.0f')
```

# Numbers Styling

```python
sns.heatmap(df2,annot=True,fmt='.0f',annot_kws= dict(size=15,weight='bold'),linewidths=0.5)
```

```python
df2.corr()
```

```python
sns.heatmap(df2.corr(),vmin=-1,vmax=1)
```

```python
sns.heatmap(df2.corr(),vmin=-1,vmax=1,annot=True,fmt='.1f')
```
