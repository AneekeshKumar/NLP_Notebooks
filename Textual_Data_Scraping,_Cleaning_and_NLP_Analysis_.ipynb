**Data Extraction Function -**

Retrieves the Title `<h1 class_='entry-title'>` and Content `<div class_='td-post-content'>` from the Blackcoffer Article Websites.



* Disposes(Decomposes) the tags.
* Cleans and Returns the text with only textual and semantic content.

````python
# Function to extract the article title and text from the given URL
def extract_article(url):
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Failed to retrieve article from: {url}")

    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract the title
    title = soup.find('h1', class_='entry-title')
    if title:
        title = title.get_text()
    else:
        title = " "

    # Extract the article text
    article_div = soup.find('div', class_='td-post-content')
    if article_div:
        # Remove any unwanted tags or content (header, footer, etc.)
        for unwanted in article_div(['header', 'footer', 'script', 'style']):
            unwanted.decompose()

        article_text = article_div.get_text(separator="\n", strip=True)
    else:
        article_text = " "

    return title, article_text
````

Creating Stop Words List

````python
def read_stop_words(file_paths):
    stop_words = set()
    encodings = ['utf-8', 'latin1', 'iso-8859-1']  #Had to try multiple encoding as the default one (utf-8) was giving UnicodeDecodeError: errors for text files.

    for file_path in file_paths:
        for encoding in encodings:
            try:
                with open(file_path, 'r', encoding=encoding) as file:
                    for line in file:
                        stop_words.add(line.strip().lower())
                break  # Break out of the encoding loop if reading is successful
            except UnicodeDecodeError:
                continue  # Try the next encoding if decoding fails

    return stop_words
````

####Cleaning using Stop Words Lists

We have a list `stop_words` now, with all your provided stop words. We can find these words in all individual articles and clear them of off these words.**bold text**

````python
def remove_stop_words(article, stopwords):
  encodings = ['utf-8', 'latin1', 'iso-8859-1'] #Had to try multiple encoding as the default one (utf-8) was giving UnicodeDecodeError: errors for text files.
  for encoding in encodings:
    try:
      with open(article, 'r', encoding=encoding) as text:
        text = text.read().lower()
      break  # Break out of the encoding loop if reading is successful
    except UnicodeDecodeError:
      continue  # Try the next encoding if decoding fails


  text = text.split()
  filtered_words = []
  for word in text:
    if word not in stopwords:
      filtered_words.append(word)
  return filtered_words
````

**Calculating Scores**

---


We convert the text into a list of tokens using the **nltk tokenize module** and use these tokens to calculate the 4 variables described below:

**Positive Score:** This score is calculated by assigning the value of +1 for each word if found in the Positive Dictionary and then adding up all the values.

**Negative Score:** This score is calculated by assigning the value of -1 for each word if found in the Negative Dictionary and then adding up all the values. We multiply the score with -1 so that the score is a positive number.

**Polarity Score:** This is the score that determines if a given text is positive or negative in nature. It is calculated by using the formula:

```
Polarity Score = (Positive Score – Negative Score)/ ((Positive Score + Negative Score) + 0.000001)
#Range is from -1 to +1
```

**Subjectivity Score:** This is the score that determines if a given text is objective or subjective. It is calculated by using the formula:
```
Subjectivity Score = (Positive Score + Negative Score)/ ((Total Words after cleaning) + 0.000001)
Range is from 0 to +1
```

````python
def calculate_scores(articles, positive_words, negative_words):
    positive_scores = {}
    negative_scores = {}
    polarity_scores = {}
    subjectivity_scores = {}

    for article_id, words in articles.items():
        # Tokenize the article
        tokens = [word.lower() for word in words]


        # Calculate Positive and Negative Scores
        positive_score = sum(1 for word in tokens if word in positive_words)
        negative_score = sum(1 for word in tokens if word in negative_words)

        # Polarity Score
        polarity_score = (positive_score - negative_score) / ((positive_score + negative_score) + 0.000001)

        # Subjectivity Score
        subjectivity_score = (positive_score + negative_score) / (len(tokens) + 0.000001)

        # Store the scores in their respective dictionaries
        positive_scores[article_id] = positive_score
        negative_scores[article_id] = negative_score
        polarity_scores[article_id] = polarity_score
        subjectivity_scores[article_id] = subjectivity_score

    return positive_scores, negative_scores, polarity_scores, subjectivity_scores


# Calculate scores for the articles
positive_scores, negative_scores, polarity_scores, subjectivity_scores = calculate_scores(articles_dict, positive_words, negative_words)

# Print the scores
print("Positive Scores:", positive_scores)
print("Negative Scores:", negative_scores)
print("Polarity Scores:", polarity_scores)
print("Subjectivity Scores:", subjectivity_scores)
````

````python
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

def tokenize_text(text):
    words = word_tokenize(text)
    sentences = sent_tokenize(text)
    return words, sentences

def calculate_average_sentence_length(words, sentences):
    return len(words) / len(sentences)

def count_syllables(word):
    token = SyllableTokenizer()
    syllables = token.tokenize(word)
    syllable_count = len(syllables)

    # Handle exceptions: words ending with "es" or "ed"
    if word.endswith('es') or word.endswith('ed'):
        if len(syllables) > 1:
            syllable_count -= 1

    return max(syllable_count, 1)
````

````python
def count_complex_words(words):
    return sum(1 for word in words if count_syllables(word) > 2)

def calculate_percentage_complex_words(complex_word_count, total_words):
    return (complex_word_count / total_words) * 100

def calculate_fog_index(avg_sentence_length, percentage_complex_words):
    return 0.4 * (avg_sentence_length + percentage_complex_words)
````

2. **Analysis of Readability**

Analysis of Readability is calculated using the Gunning Fox index formula described below.

**Average Sentence Length**

`Average Sentence Length = the number of words / the number of sentences`

**Percentage of Complex words**

`Percentage of Complex words = the number of complex words / the number of words `

**Fog Index**

`Fog Index = 0.4 * (Average Sentence Length + Percentage of Complex words)`

---

3.**Average Number of Words Per Sentence**

The formula for calculating is:

`Average Number of Words Per Sentence = the total number of words / the total number of sentences`

---

4. **Complex Word Count**

Complex words are words in the text that ***contain more than two syllables.***

---

6. Syllable Count Per Word

We count the number of Syllables in each word of the text by counting the vowels present in each word. We also handle some exceptions like words ending with "es","ed" by not counting them as a syllable.

---
---

````python
def calculate_readability_metrics(file_path):
    stop_words = set(stopwords.words('english'))
    sentences = []
    tokens = []
    total_words = 0
    complex_word_count = 0
    syllables_per_word = {}

    with open(file_path, 'r') as file:
        for line in file:
            # Tokenize the line into sentences
            line_sentences = sent_tokenize(line)
            sentences.extend(line_sentences)

            # Tokenize the line into words
            line_tokens = word_tokenize(line)
            line_tokens = [word.lower() for word in line_tokens if word.isalpha()]

            # Remove stop words
            cleaned_tokens = [word for word in line_tokens if word not in stop_words]
            tokens.extend(cleaned_tokens)

    # Calculate total words after cleaning
    total_words = len(tokens)

    # Calculate average sentence length
    avg_sentence_length = total_words / len(sentences) if len(sentences) > 0 else 0

    # Count complex words
    complex_word_count = count_complex_words(tokens)

    # Calculate percentage of complex words
    percentage_complex_words = (complex_word_count / total_words) * 100 if total_words > 0 else 0

    # Calculate Fog Index
    fog_index = 0.4 * (avg_sentence_length + percentage_complex_words)

    # Calculate syllables per word
    for word in tokens:
        syllables_per_word[word] = count_syllables(word)

    return {
        'average_sentence_length': avg_sentence_length,
        'percentage_complex_words': percentage_complex_words,
        'fog_index': fog_index,
        'complex_word_count': complex_word_count,
        'avg_words_per_sentence': avg_sentence_length,
        'syllables_per_word': syllables_per_word,
        'word_count': total_words
    }
````

---
**Personal Pronouns**

To calculate Personal Pronouns mentioned in the text, we use regex to find the counts of the words - `“I,” “we,” “my,” “ours,” and “us”`. Special care is taken so that the country name `US` is not included in the list.

---

````python
def count_personal_pronouns(file_path):
    pronouns = ["I", "we", "my", "ours", "us"]
    pronoun_counts = {pronoun.lower(): 0 for pronoun in pronouns}  # Convert keys to lowercase

    # Read text from file
    with open(file_path, 'r') as file:
        text = file.read()

    # Regex pattern to find occurrences of pronouns as whole words (case insensitive)
    pattern = r'\b(?:{})\b'.format('|'.join(pronouns))

    # Regex pattern to exclude "US" specifically
    exclude_pattern = r'\bUS\b'

    # Find all matches using regex (case insensitive)
    matches = re.findall(pattern, text, flags=re.IGNORECASE)
    exclude_matches = re.findall(exclude_pattern, text, flags=re.IGNORECASE)

    # Count occurrences for each pronoun
    for match in matches:
        match_lower = match.lower()  # Convert match to lowercase
        if match_lower not in exclude_matches:
            pronoun_counts[match_lower] += 1

    return pronoun_counts
````

---

**Average Word Length**

Average Word Length:
`Sum of the total number of characters in each word/Total number of words`

---

````python
def calculate_avg_word_length(file_path):
    total_chars = 0
    total_words = 0

    # Read text from file
    with open(file_path, 'r') as file:
        text = file.read()

    # Tokenize text into words
    words = text.split()

    # Calculate total characters and total words
    total_chars = sum(len(word) for word in words)
    total_words = len(words)
    # print(total_chars, total_words)
    # Calculate average word length
    if total_words > 0:
        avg_word_length = total_chars / total_words
    else:
        avg_word_length = 0

    return avg_word_length
````
